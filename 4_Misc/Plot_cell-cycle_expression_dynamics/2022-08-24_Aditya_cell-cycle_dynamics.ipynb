```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt,cm
from matplotlib.gridspec import GridSpec as gs
import pickle as pk
from scipy.stats import mannwhitneyu
```

```python
adi = pd.read_excel('Aditya-1.xlsx')
```

```python
time = np.array([float(x[:-3]) for x in adi.columns[4:].values])
resr = adi[adi['ORF ID']=='Rv1830'].values[0][4:]
whiB2 = adi[adi['ORF ID']=='Rv3260c'].values[0][4:]
fig=plt.figure(figsize=(4,3))
plt.plot(time,resr,label='$resR$',lw=3)
plt.plot(time,whiB2,label='$whiB2$',lw=3)
plt.legend(prop={'size':12,'family':'Arial'})
plt.xticks([10,20,30,40,50],[10,20,30,40,50],fontsize=14,fontname='Arial')
plt.yticks([0.5,1.0,1.5,2.0],[0.5,1.0,1.5,2.0],fontsize=14,fontname='Arial')
plt.xlabel('Hours since\nreplication resumption',fontsize=14,fontname='Arial')
plt.ylabel('Normalized expression',fontsize=14,fontname='Arial')
plt.title('Cell-cycle dependent gene expression\n(Aditya et. al., 2020)',fontsize=14,fontname='Arial')
for i in np.linspace(22,37,30):
    x1=i
    x2=i+16/31
    a = 1-(abs(np.mean([x1,x2])-30)/8)
    if a>1:
        a=1
    color = cm.get_cmap('Greys')(a)
    plt.fill_betweenx(y=[0,2.2],x1=x1,x2=x2,alpha=0.25,
                      color=color,ec=None)

plt.text(30,0.2,'Cytokinesis',ha='center',fontsize=14,fontname='Arial',va='bottom')
plt.ylim(0.0,2.2)
plt.savefig('/Users/jz-rolling/Desktop/Rv1830_manuscript/figures/cell_cycle_expression.png',
           bbox_inches='tight',dpi=160)
```
